```python
text_generator = pipeline("text-generation", model="gpt2")
```

```python
# Generate text based on a prompt
prompts = [
    "Explain about wichita state university?",
    "Explain the concept of Supervised Learning",
    "Tell a short story about a robot in the future.",
    "Describe about international student life in USA",
    "World War-2 got ended in the year 1945. True or False?",
    "Create a dialogue between a teacher and a student about Paper presentation",
    "Write an article about the importance of mental health.",
    "Compose an Indian song about love",
    "Generate a motivational quote for the students",
    "Produce a Jhony Jhony yes papa rhyme"
]
```

```python
results = []
for prompt in prompts:
    generated_text = text_generator(prompt, max_length=100, num_return_sequences=1)
    results.append((prompt, generated_text[0]['generated_text']))
```

```python
for prompt, text in results:
    print(f"Prompt: {prompt}\nGenerated Text: {text}\n")
```
